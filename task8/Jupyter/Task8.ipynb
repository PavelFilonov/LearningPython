```python
import numpy as np

x = np.linspace(0, 10, 501)
ampl = np.exp(-(x - 3.5)**2 / 0.8)
y = np.sin(x * 25) * ampl

#y = np.sin(x) * (x + 1)

import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(x, y, label='signal')
plt.xlabel('time')
plt.ylabel('amplitude')
plt.legend()
```

```python
import numpy as np

x = np.linspace(0, 10, 501)
y = np.sin(x * (x + 0.1))

import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(x, y, label='signal')
plt.xlabel('time')
plt.ylabel('amplitude')
plt.legend()
```
